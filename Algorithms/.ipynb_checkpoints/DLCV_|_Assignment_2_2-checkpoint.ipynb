```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
from PIL import Image
from scipy.io import loadmat

np.random.seed(7)
```

```python
def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))

def tanh(Z):
    return np.tanh(Z)

def ReLU(Z):
    return Z * (Z > 0)
```

```python
# sigmoid derivative
def sigmoid_back(Z):
    return sigmoid(Z) * (1 - sigmoid(Z))

# tanh derivative
def tanh_back(Z):
    return 1 - np.power(np.tanh(Z),2)

# relu derivative
def ReLU_back(Z):
    return (Z > 0) * 1
```

```python
# cross_entropy cost
def cross_entropy(Y_hat, Y):
    m = Y_hat.shape[1]
    cost = -1 / m * (np.dot(Y.T, np.log(Y_hat)) + np.dot((1 - Y).T, np.log(1 - Y_hat)))
    return np.squeeze(cost)
```

```python
class NEURON:
  def __init__(self, input_length, w=None, b=0, activation='Linear'):
    self.activation = activation
    
    if(not w):
      w = np.random.rand(input_length, ) * np.sqrt(1. / input_length)

    self.w = w

    if(not b):
      b = 0
      
    self.b = b

  def forward(self, X):
    self.Z = np.dot(X, self.w) + self.b
    
    if self.activation == 'Linear':
      self.A = self.Z

    elif self.activation == 'Sigmoid':
      self.A = sigmoid(self.Z)

    elif self.activation == 'ReLU':
      self.A = ReLU(self.Z)
    
    elif self.activation == 'Tanh':
      self.A = tanh(self.Z)

    return self.Z, self.A

  def backprop(self, dw, db):
    self.dw = dw
    self.db = db

  def update(self, lr):
    assert(self.w.shape == self.dw.shape)

    self.w = self.w - lr * self.dw
    self.b = self.b - lr * self.db

  def print_neuron(self):
    print(f"\t\tWeights: {self.w}")
    print(f"\t\tBiases: {self.b}")
```

```python
class INPUT_LAYER:
  def __init__(self, X_length):
    self.X_length = X_length

  def forward(self, X, eps=None):
    assert(X.shape[1] == self.X_length)
    return X

  def print_layer(self):
      print(f"Input Shape: {self.X_length}")
      print("")
```

```python
class HIDDEN_LAYER:
  def __init__(self, input_length, output_length, W=None, b=None, activation='ReLU'):
    self.activation = activation
    self.b = b
    self.neurons = [] 

    for i in range(output_length):
      w_n = None
      b_n = None

      if(W):
        w_n = W[:, i]
      if(b):
        b_n = b[:, i].squeeze()

      self.neurons.append(NEURON(input_length, activation=activation, w=w_n, b=b_n))

  def forward(self, X):
    self.X = X
    self.A = []
    self.Z = []

    for i, neuron in enumerate(self.neurons):
      assert(X.shape[1] == neuron.w.shape[0])
      Z, A = neuron.forward(X)
      self.Z.append(Z)
      self.A.append(A)

    self.A = np.array(self.A).T
    self.Z = np.array(self.Z).T

    return self.Z, self.A

  def backprop(self, dZ_next, W_next):
    m = self.A.shape[0]
    
    dA = np.dot(dZ_next, W_next.T)

    if self.activation == 'ReLU':
      dZ = dA * ReLU_back(self.Z)
    elif self.activation == 'Sigmoid':
      dZ = dA * sigmoid_back(self.Z)
    elif self.activation == 'Tanh':
      dZ = dA * tanh_back(self.Z)
      
    dW = (1 / m) * np.dot(self.X.T, dZ)
    db = (1 / m) * np.sum(dZ, axis=0, keepdims=True)

    assert(dW.shape == (self.X.shape[1], len(self.neurons)))

    W = []
    for i, neuron in enumerate(self.neurons):
      neuron.backprop(dW[:, i], db[:, i].squeeze())
      W.append(neuron.w)

    return dZ, np.array(W).T

  def update(self, lr):
    for neuron in self.neurons:
      neuron.update(lr)

  def print_layer(self):
    for i, neuron in enumerate(self.neurons):
      print(f"\tNeuron {i}: ")
      neuron.print_neuron()
      print("")
```

```python
class OUTPUT_LAYER:
  def __init__(self, input_length, output_length, W=None, b=None, activation='Sigmoid'):
    self.activation = activation
    self.b = b
    self.neurons = [] 

    for i in range(output_length):
      w_n = None
      b_n = None

      if(W):
        w_n = W[:, i]
      if(b):
        b_n = b[:, i].squeeze()

      self.neurons.append(NEURON(input_length, activation=activation, w=w_n, b=b_n))

  def forward(self, X):
    self.X = X
    self.A = []
    self.Z = []

    for i, neuron in enumerate(self.neurons):
      assert(X.shape[1] == neuron.w.shape[0])   
      Z, A = neuron.forward(X)
      self.Z.append(Z)
      self.A.append(A)
    
    self.A = np.array(self.A).T
    self.Z = np.array(self.Z).T

    return self.Z, self.A

  def backprop(self, y):
    m = self.A.shape[0]

    dZ = (self.A - y)
    dW = (1 / m) * np.dot(self.X.T, dZ)
    db = (1 / m) * np.sum(dZ, axis=0, keepdims=True)

    assert(dW.shape == (self.X.shape[1], len(self.neurons)))
    
    W = []
    for i, neuron in enumerate(self.neurons):
      neuron.backprop(dW[:, i], db[:, i].squeeze())
      W.append(neuron.w)

    return dZ, np.array(W).T

  def update(self, lr):
    for neuron in self.neurons:
      neuron.update(lr)

  def print_layer(self):
    for i, neuron in enumerate(self.neurons):
      print(f"\tNeuron {i}: ")
      neuron.print_neuron()
```

```python
class NN_MLP:
  def __init__(self, layers, activations, W=None, b=None):
    self.layers = []

    for i in range(len(layers)):
      w_n = None
      b_n = None

      if(W):
        w_n = W[i - 1]
      if(b):
        b_n = b[i - 1]

      if(i == 0):
        self.layers.append(INPUT_LAYER(layers[i]))

      elif(i == len(layers) - 1):
          self.layers.append(OUTPUT_LAYER(layers[i - 1], layers[i], activation=activations[i], W=w_n, b=b_n))
      else:
          self.layers.append(HIDDEN_LAYER(layers[i - 1], layers[i], activation=activations[i], W=w_n, b=b_n))

  def forward(self, X):
    A = X

    for i, layer in enumerate(self.layers):
      
      if(type(layer) is INPUT_LAYER):
        Res = layer.forward(A)
      else:
        _, Res = layer.forward(A)

      assert(A.shape[0] == Res.shape[0])

      if(type(layer) is not INPUT_LAYER):
        assert(Res.shape[1] == len(layer.neurons))

      A = Res

    return A

  def cost(self, A, y):
    return cross_entropy(A, y)

  def backprop(self, y):
    dZ, W = (None, None)
    for layer in reversed(self.layers):
      if(type(layer) is OUTPUT_LAYER):
        dZ, W = layer.backprop(y)
        
      elif(type(layer) is HIDDEN_LAYER):
        dZ, W = layer.backprop(dZ, W)
        
  def update(self,lr):
    for layer in self.layers:
      if(type(layer) is not INPUT_LAYER):
        layer.update(lr)

  def train(self, X, y, lr=1e-5, epochs=30):
    costs = []
    epoch_nums = []
    for epoch in range(epochs):
      epoch_nums.append(epoch)

      A = self.forward(X)

      cur_cost = self.cost(A, y) 
      costs.append(cur_cost)

      self.backprop(y)
      self.update(lr)
    
    plt.title("Error vs Epochs")
    plt.xlabel('Number of Epochs') 
    plt.ylabel('Error') 
    plt.plot(epoch_nums, costs)

    return costs


  def print_net(self):
    for i, layer in enumerate(self.layers):
      if(type(layer) is INPUT_LAYER):
        print(f"<<<Input Layer>>>")
        layer.print_layer()
      elif(type(layer) is HIDDEN_LAYER):
        print(f"<<<Hidden Layer {i}>>>")
        layer.print_layer()
      elif(type(layer) is OUTPUT_LAYER):
        print(f"<<<Output Layer>>>")
        layer.print_layer()
```

```python
layers = [30, 20, 10, 4, 1]
activations = ['-', 'ReLU', 'ReLU', 'ReLU', 'Sigmoid']
network = NN_MLP(layers, activations)
```
